```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
nome_arquivo = 'NumeroReservasPorHotel/part-r-00000.txt'

# Listas para armazenar os IDs dos hotéis e o número de reservas
ids_hoteis = []
num_reservas = []

# Ler os dados do arquivo
with open(nome_arquivo, 'r') as arquivo:
    linhas = arquivo.readlines()
    for linha in linhas:
        id_hotel, reservas = linha.strip().split('\t')
        ids_hoteis.append(int(id_hotel))
        num_reservas.append(int(reservas))

# Ordenar os hotéis com base no número de reservas em ordem decrescente
ids_hoteis, num_reservas = zip(*sorted(zip(ids_hoteis, num_reservas), key=lambda x: x[1], reverse=True))

# Selecionar os top 10 hotéis com mais reservas
top_10_ids = ids_hoteis[:10]
top_10_reservas = num_reservas[:10]

# Criar o gráfico de pontos (scatter plot)
plt.scatter(top_10_ids, top_10_reservas, s=100, c='red', alpha=0.8)
plt.xlabel('ID do Hotel')
plt.ylabel('Número de Reservas')
plt.title('Top 10 Hotéis com Mais Reservas')

# Adicionar rótulos aos pontos
for i, id_hotel in enumerate(top_10_ids):
    plt.text(id_hotel, top_10_reservas[i], str(id_hotel), ha='center', va='bottom')

# Mostrar o gráfico
plt.show()
```

```python
nome_arquivo = 'NightsPerOriginAverage/part-r-00000.txt'

# Listas para armazenar as origens e médias de noites
origens = []
medias_noites = []

# Ler os dados do arquivo
with open(nome_arquivo, 'r') as arquivo:
    linhas = arquivo.readlines()
    for linha in linhas:
        origem, media_noites = linha.strip().split('\t')
        origens.append(origem)
        medias_noites.append(float(media_noites))

# Ordenar as origens e médias de noites com base nas médias em ordem decrescente
origens, medias_noites = zip(*sorted(zip(origens, medias_noites), key=lambda x: x[1], reverse=True))

# Selecionar as top 10 origens e médias de noites
top_10_origens = origens[:10]
top_10_medias_noites = medias_noites[:10]

# Criar o gráfico de barras horizontais
plt.barh(top_10_origens, top_10_medias_noites)
plt.xlabel('Média de Noites')
plt.ylabel('Origens')
plt.title('Principais Origens com Média de Noites')

# Mostrar o gráfico
plt.show()
```

```python
# Ler os dados do arquivo .txt
data = []
with open('PrecoMedioPorNoitePorHotel/part-r-00000.txt', 'r') as file:
    for line in file:
        hotel_id, preco_por_noite = line.split()
        data.append((int(hotel_id), float(preco_por_noite)))

# Ordenar os dados pelo preço por noite em ordem decrescente
data.sort(key=lambda x: x[1], reverse=True)

# Selecionar os top 5 hotéis mais caros
top_5_caros = data[:5]

# Selecionar os 5 hotéis mais baratos
top_5_baratos = data[-5:]

# Juntar os dados dos top 5 hotéis mais caros e dos 5 hotéis mais baratos
top_10 = top_5_caros + top_5_baratos

# Separar os dados em listas separadas para o gráfico
hotel_ids = [item[0] for item in top_10]
precos = [item[1] for item in top_10]
nomes_hoteis = ['Hotel {}'.format(hotel_id) for hotel_id in hotel_ids]

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))  # Ajustar o tamanho da figura
plt.bar(nomes_hoteis, precos, color=['green'] * 5 + ['red'] * 5)  # Adicionar cor às barras

plt.xlabel('Hotel')
plt.ylabel('Preço por Noite')
plt.title('Top 5 Mais Caros e Top 5 Mais Baratos')

# Rotacionar os rótulos dos hotéis para melhorar a legibilidade
plt.xticks(rotation=45, ha='right')

# Adicionar valores acima das barras
for i, preco in enumerate(precos):
    plt.text(i, preco, str(preco), ha='center', va='bottom')

plt.tight_layout()  # Ajustar o espaçamento entre as barras
plt.show()

```
