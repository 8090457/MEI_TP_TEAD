```python
import pandas as pd
```

```python
df = pd.read_csv('../Datasets/dataSetFinalML.csv', delimiter=';')

df.describe()
```

```python
# drop unnecessary columns

columns_to_drop = [
    "Q_tipo_quarto",
    "rate_plan",
    "capacidade_maxima",
    "capacidade_max_criancas",
    "capacidade_max_bebes",
    "localizacao",
    "estrelas",
    "hora_max_checkin",
    "is_holiday",
    "event_count",
    "temperature_avg",
    "temperature_max",
    "temperature_min",
    "preco_euros",
    "criancas",
    "bebes",
    "quantidade"
]

df = df.drop(columns_to_drop, axis=1)

df.dtypes
```

```python
df['data_chegada'] = pd.to_datetime(df['data_chegada'])
df['data_partida'] = pd.to_datetime(df['data_partida'])

df['mes_chegada'] = df['data_chegada'].dt.month
df['mes_partida'] = df['data_partida'].dt.month
df
```
