```python
# Este notebook cria um dataset com a ocupação por dia, para um hotel/mês específico
# A ideia é fazer-se algo parecido, para todos os hoteis e por mês

import pandas as pd
import numpy as np
```

```python
xls = pd.ExcelFile('e-GDS-v3.xlsx')
reservas = pd.read_excel(xls, 'QuartosReservados')
```

```python
dates = reservas['Data chegada'].str.split('-', expand=True)
dates
```

```python
reservas['Ano'] = dates[0]
reservas['Mes'] = dates[1]
reservas['Room ID'] = reservas['Room ID'].astype(str)
```

```python
total_quartos = 5
r_03 = reservas.loc[(reservas['Hotel ID'] == 493) & (reservas['Ano'] == '2022') & (reservas['Mes'] == '03')]
```

```python
marco = pd.DataFrame(pd.date_range(start='2022-03-01', end='2022-08-31'), columns=['date'])
marco
```

```python
joined = marco.set_index('date').join(r_03.set_index('Data chegada'))
joined = joined.fillna(0)
joined['Data chegada'] = joined.index
joined
```

```python
table = pd.pivot_table(joined, values='Número de noites', index=['Data chegada'],columns=['Room ID'], aggfunc=np.sum, fill_value=0).reset_index()
table
```

```python
lista1 = table['2324'].to_numpy()
for index in range(len(lista1)):
    if index > 0:
        if lista1[index-1] > 0:
            lista1[index] = lista1[index-1] - 1

lista2 = table['2338'].to_numpy()
for index in range(len(lista2)):
    if index > 0:
        if lista2[index-1] > 0:
            lista2[index] = lista2[index-1] - 1

lista3 = table['2800'].to_numpy()
for index in range(len(lista3)):
    if index > 0:
        if lista3[index-1] > 0:
            lista3[index] = lista3[index-1] - 1

zeros = arr0 = [0] * len(lista1)
```

```python
table['2324'] = (lista1 != zeros).astype(int)
table['2338'] = (lista2 != zeros).astype(int)
table['2800'] = (lista3 != zeros).astype(int)
```

```python
table['occupation'] = (table['2324'] + table['2338'] + table['2800'])/5
table
```
